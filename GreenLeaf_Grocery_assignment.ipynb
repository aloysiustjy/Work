```python
rice_bags = 50
milk_cartons = 40
snack_packs = 100
store_name = "GreenLeaf Grocery"
store_open = True
prices = {
    "rice": 10.0,
    "milk": 2.50,
    "snack": 1.50
}
print("Please Enter you daily Sales Numbers")
rice_sold = int(input("Rice bags sold today: "))
milk_sold = int(input("Milk cartons sold today: "))
snacks_sold = int(input("Snack packs sold today: "))
rice_bags -= rice_sold
milk_cartons -= milk_sold
snack_packs -= snacks_sold
rice_revenue = rice_sold * prices["rice"]
milk_revenue = milk_sold * prices["milk"]
snack_revenue = snacks_sold * prices["snack"]
total_revenue = rice_revenue + milk_revenue + snack_revenue
made_over_300 = total_revenue > 300
low_stock_threshold = 10
urgent_restock = (rice_bags < 5) or (milk_cartons < 5)
rice_low = rice_bags < low_stock_threshold
milk_low = milk_cartons < low_stock_threshold
snacks_low = snack_packs < low_stock_threshold
print("Please enter 5 Customer Comments")
customer_comments = []
customer_comments.append(input("Comment 1: "))
customer_comments.append(input("Comment 2: "))
customer_comments.append(input("Comment 3: "))
customer_comments.append(input("Comment 4: "))
customer_comments.append(input("Comment 5: "))
store_info = {
    "address": "78 Pine Avenue",
    "city": "Riverview",
    "phone": "555-3098"
}
print("         GREENLEAF GROCERY - DAILY REPORT")
print("===================================================\n")
print("----- INVENTORY REMAINING -----")
print(f"Rice bags:   {rice_bags}")
print(f"Milk cartons:{milk_cartons}")
print(f"Snack packs: {snack_packs}\n")
print("---------- REVENUE ----------")
print(f"Rice revenue:  ${rice_revenue:.2f}")
print(f"Milk revenue:  ${milk_revenue:.2f}")
print(f"Snack revenue: ${snack_revenue:.2f}")
print(f"TOTAL:         ${total_revenue:.2f}")
if made_over_300:
    print("Great! Sales exceeded $300 today.\n")
else:
    print("Sales did not reach $300.\n")
    print("------ STOCK WARNINGS ---c---")
if rice_low or milk_low or snacks_low:
    if rice_low:
        print("- Rice is running low.")
    if milk_low:
        print("- Milk is running low.")
    if snacks_low:
        print("- Snacks are running low.")
else:
    print("All items are sufficiently stocked.")
if urgent_restock:
  print("URGENT: Rice or milk below 5 units, reorder immediately!")
print("----- CUSTOMER FEEDBACK -----")
for comment in customer_comments:
    print(comment)
print("\n----- STORE DETAILS -----")
print(f"Store Name: {store_name}")
print(f"Address:    {store_info['address']}")
print(f"City:       {store_info['city']}")
print(f"Phone:      {store_info['phone']}")
print(f"Status:     {'Open' if store_open else 'Closed'}")
print("\n===================================================")
print("                END OF DAILY REPORT")
```
