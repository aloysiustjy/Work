```python
gym_name = "SmartFit Gym"
is_open = True

print("Welcome to", gym_name)
print("Gym open status:", is_open)
members = []
print("Add 5 sample members checking in:")
while len(members) < 5:
    name = input(f"Enter member name #{len(members)+1}: ").strip()
    if name:
        members.append(name)
    else:
        print("Error: Name cannot be empty. Try again.")
print("Current checked-in members:", members)
print("--------------------------------------")
equipment = {
    "treadmills": {"total": 8, "in_use": 2},
    "bikes": {"total": 5, "in_use": 1},
    "rowing_machines": {"total": 4, "in_use": 0},
    "dumbbells": {"total": 20, "in_use": 6}
}
print("Initial equipment status:")
print(equipment)
print("--------------------------------------")
equipment["treadmills"]["in_use"] += 3
equipment["dumbbells"]["in_use"] += 2
while True:
    print("\nCurrent equipment:", list(equipment.keys()))
    item = input("Which equipment do you want to update? ").strip()
    if item in equipment:
        amount = input(f"How many more {item} are now in use? ")
        if amount.isdigit():
            equipment[item]["in_use"] += int(amount)
        else:
            print("Invalid input. Defaulting to +1.")
            equipment[item]["in_use"] += 1
    else:
        print("Invalid equipment name. Try again.")
        continue
    more = input("Update anything else? (yes/no): ").strip().lower()
    if more not in ["yes", "y"]:
        break
print("\nUpdated equipment usage:")
print(equipment)
low_items = []
fully_occupied = []
most_used_name = None
most_used_amount = -1
for item in equipment:
  total = equipment[item]["total"]
  used = equipment[item]["in_use"]
  remaining = total - used
  if used > most_used_amount:
    most_used_amount = used
    most_used_name = item
  if remaining < 2:
    low_items.append(item)
  if remaining <= 0:
    fully_occupied.append(item)
print("========== SMARTFIT GYM – DAILY SUMMARY ==========")
print("Gym Name:", gym_name)
print("Members Who Checked In Today:")
print(members)
print("\nFinal Equipment Comparison Status (Total vs In Use):")
for item in equipment:
    t = equipment[item]["total"]
    u = equipment[item]["in_use"]
    print(f" - {item}: {u} / {t} in use")
print("\nLow Availability Items (<2 remaining):")
print(low_items if low_items else "None")
print("\nEquipment Fully Occupied:")
print(fully_occupied if fully_occupied else "None")
print("\nMost Used Equipment Today:")
print(most_used_name, "with", most_used_amount, "units in use")
```
