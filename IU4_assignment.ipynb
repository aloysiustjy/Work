```python
class Product:
  def __init__(self,name,price,quantity):
    self.name = name
    self.price = price
    self.quantity = quantity
  def update_stock(self):
    new_items = int(input("Enter the amount of new items: "))
    used_items = int(input("Enter the amount of items used or sold: "))
    self.quantity += new_items - used_items
  def total_value(self):
    return self.price * self.quantity

product_1_name = input("Enter the name of your first product: ")
product_1_price = float((input("Enter the price of product 1: ")))
product_1_quantity = int((input("Enter the quantity of product 1: ")))
product_2_name = (input("Enter the name of product 2: "))
product_2_price = float((input("Enter the price of product 2: ")))
product_2_quantity = int((input("Enter the quantity of product 2: ")))
product_1 = Product(product_1_name, product_1_price, product_1_quantity)
product_2 = Product(product_2_name, product_2_price, product_2_quantity)

product_1.update_stock()
product_2.update_stock()
product_1.total_value()
product_2.total_value()
print(f"{product_1.name} - Quantity: {product_1.quantity}, Total value: {product_1.total_value()}")
print(f"{product_2.name} - Quantity: {product_2.quantity}, Total value: {product_2.total_value()}")
```
