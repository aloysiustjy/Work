# Inventory Management:
You run a small store and have the following inventory:
{"apples": 50, "oranges": 100, "bananas": 150}.
A customer buys 10 apples, 20 oranges, and 15 bananas. Update the inventory.

# Student Grades:
You have the following dictionary with student names and their grades:
{"Alice": 85, "Bob": 78, "Charlie": 92}.
Add a new student "David" with a grade of 88.

# Phonebook:
You have a phonebook dictionary with names as keys and phone numbers as values:
{"John": "123-4567", "Jane": "234-5678", "Doe": "345-6789"}.
Retrieve Jane's phone number.

# City Population:
Given the population of cities:
{"New York": 8000000, "Los Angeles": 4000000, "Chicago": 2700000},
check if "Houston" is in the dictionary.

# Word Frequency:
Given a list of words ["apple", "banana", "apple", "orange", "banana", "apple"],
create a dictionary to count the frequency of each word.

# Update Prices:
You have a dictionary with product prices:
{"milk": 1.2, "bread": 2.5, "eggs": 3.0}.
The price of bread increases by 0.5. Update the dictionary.

# Merge Data:
You have two dictionaries with employee details:
{"name": "John", "age": 30} and {"department": "HR", "salary": 5000}.
Merge these dictionaries into one.

# Book Collection:
You have a dictionary with book titles and their authors:
{"1984": "George Orwell", "To Kill a Mockingbird": "Harper Lee", "The Great Gatsby": "F. Scott Fitzgerald"}.
Remove the book "1984" from the collection.

# Country Capitals:
Given a dictionary of countries and their capitals:
{"USA": "Washington, D.C.", "France": "Paris", "Japan": "Tokyo"},
add a new country "India" with its capital "New Delhi".

# Translate Dictionary:
You have a dictionary that translates English words to Spanish:
{"hello": "hola", "world": "mundo", "friend": "amigo"}.
Translate the sentence "hello world" using the dictionary.
